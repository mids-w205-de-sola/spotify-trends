```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('final_data.csv')
df
```

```python
df.describe()
```

```python
df.groupby(['date']).mean()
```

```python
Does the valence (happiness) of songs increase as we approach Christmas?

Response: While valence does increase a bit post Thanksgiving, it does not noticeably change during the holidays.
```

```python
import seaborn as sns

plt.figure(figsize=(12,8))
ax = sns.lineplot(x='date', y='valence', data=df).set(title='Valence by Date', xlabel = 'Date', ylabel = 'Valence')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(df.date, df.valence, color = 'blue', width = .5, alpha = 1)

plt.xlabel("Date", fontsize=10)
plt.ylim([0,1])
plt.xticks(rotation='vertical')
plt.ylabel("Valence",fontsize=12)
plt.title('Valence by Date',fontsize=14)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(df.date, df.danceability, color = 'blue', width = .5, alpha = 1)

plt.xlabel('Date', fontsize=10)
plt.xticks(rotation='vertical')
plt.ylabel('Danceability',fontsize=12)
plt.title('Danceability by Date',fontsize=14)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(df.date, df.energy, color = 'blue', width = .5, alpha = 1)

plt.xlabel('Date', fontsize=10)
plt.xticks(rotation='vertical')
plt.ylabel('Energy',fontsize=12)
plt.title('Energy by Date',fontsize=14)
plt.show()
```
